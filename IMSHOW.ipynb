```
import matplotlib.pyplot as plt
```

```
import matplotlib.image as mpimg
```

### Reading Image using Matplotlib

```
img = mpimg.imread('/content/1(5).jpg')
```

```
plt.imshow(img)
plt.show()
```

### Reading Image using OpenCV

```
import cv2
```

```
imgcv2 = cv2.imread('/content/1(5).jpg')
```

```
plt.imshow(imgcv2)
plt.show()
```

```
imgcv2 = cv2.cvtColor(imgcv2, cv2.COLOR_BGR2RGB)
```

```
plt.imshow(imgcv2)
plt.show()
```

### Changing Aspect

```
plt.imshow(img, aspect = 0.5)
plt.show()
```

```
img1 = mpimg.imread('/content/1(3).jpg')
img2 = mpimg.imread('/content/1(4).jpg')
img3 = mpimg.imread('/content/1(5).jpg')
img4 = mpimg.imread('/content/1(6).jpg')
```

```
plt.figure()
plt.subplot(2,2,1)
plt.imshow(img1)

plt.subplot(2,2,2)
plt.imshow(img2)

plt.subplot(2,2,3)
plt.imshow(img3)

plt.subplot(2,2,4)
plt.imshow(img4)

plt.show()
```

### Color Bar

```
plt.imshow(img, aspect = 1.5)
plt.colorbar()
plt.show()
```

```
plt.imshow(img, cmap='YlOrRd')
plt.colorbar()
plt.show()
```
